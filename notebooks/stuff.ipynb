```python
import numpy as np
import cPickle
from PIL import Image
```

```python
img = '/media/Data/flipvanrijn/datasets/coco/images/train/COCO_train2014_000000000009.jpg'
```

```python
d = Image.open(img)
```

```python
data = {'pixels': d.tobytes(), 'size': d.size, 'mode': d.mode}
```

```python
pickled = cPickle.dumps(data)
```

```python
unpickled = cPickle.loads(pickled)
```

```python
Image.frombytes(unpickled['mode'], unpickled['size'], unpickled['pixels'])
```
