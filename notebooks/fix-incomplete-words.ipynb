```python
from lxml import etree
from nltk.tokenize import wordpunct_tokenize, regexp_tokenize
```

```python
filename = '/media/Data/VrijdagMiddagOpdracht/304392.xml'
with open(filename, 'r') as f:
    tree = etree.parse(f)

namespace = tree.getroot().tag[1:].split('}')[0]
```

```python
paragraphs = tree.findall('.//{%s}p' % namespace)

def bigrams(input_list):
    return zip(input_list, input_list[1:])

def valid_bigram(bigram):
    return 'qqq' not in bigram[0].lower() and 'qqq' not in bigram[1].lower()

def lookup(bigrams, first=None, second=None):
    if first is None and second is None:
        return
    idx = 0 if first is not None else 1
    val = first if first is not None else second
    
    result = []    
    for bigram in bigrams:
        if bigram[idx] == val:
            result.append(bigram)
    return result
```

```python
invalid_paragraphs = []
ngrams = []
for paragraph in paragraphs:
    text = ''.join([t for t in paragraph.itertext()])
    if 'qqq' in text.lower():
        invalid_paragraphs.append(text)
    [ngrams.append(bigram) for bigram in bigrams(wordpunct_tokenize(text)) if valid_bigram(bigram)]
print 'Num. ngrams: ', len(ngrams)
print 'Num. invalid paragraphs: ', len(invalid_paragraphs)
```

```python
unique_ngrams = list(set(ngrams))
print 'Unique ngrams: ', len(unique_ngrams)
```
